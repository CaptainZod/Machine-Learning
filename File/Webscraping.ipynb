```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.bbc.com/news'
response = requests.get(url)
response.status_code
```

```python
print(response.content)
```

```python
soup =  BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
soup.title
```

```python
headlines = soup.find('body').find_all('h2')
paragraphs = soup.find('body').find_all('p')
for x,y in zip(headlines,paragraphs):
    print("\033[1m" + x.text.strip() + "\033[0m")
    print(f"{y.text.strip()}\n")
```

```python
link = soup.find('body').find_all('a', href=True)
for l in link:
    if l['href'].startswith('/news/'):
        print(f"https://www.bbc.com{l['href']}")
    elif l['href'].startswith('/'):
        print(f"https://www.bbc.com{l['href']}")
    else:
        print(l['href'])
```

```python
img = soup.find('body').find_all('img')
for i in img:
    print(i['src'])
```

```python
#most read
read = soup.find('div',{"class":"sc-93223220-0 lbLNTF"})
print(read.prettify())
```

```python
watch = soup.find_all("div",{"class":"sc-c6f6255e-0 eGcloy"})
print(watch)
```

```python
for most_watch in watch:
    headings = most_watch.find_all("h2",{"class":"sc-87075214-3 hFNjzK"})
    links = most_watch('a')
    for heading,link in zip(headings,links):
        h_text = heading.text.strip()
        link_text = link['href']
        print(h_text)
        print(f"https://www.bbc.com{link_text}")
        print("\n")
```
