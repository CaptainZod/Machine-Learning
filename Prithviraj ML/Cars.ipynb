```python
Correct_Mileage= []
for i in df.Mileage:
    if str(i).endswith('km/kg'):
        i = i[:-6]
        i = float(i)*1.40
        Correct_Mileage.append(float(i))
 
    elif str(i).endswith('kmpl'):
        i = i[:-6]
        #print(i)
        Correct_Mileage.append(float(i))
    else:
        Correct_Mileage.append(np.nan)  # Append NaN for unrecognized format
   
df['Mileage']=Correct_Mileage
df
```

```python
msno.matrix(df)
```

```python
def convert_engine_to_numeric(engine_value):
    try:
        # Remove the 'L' and convert to float
        return float(engine_value.replace('L', ''))
    except (ValueError, AttributeError):  # Handle errors
        return np.nan  # Return NaN for invalid conversions

# Apply the conversion to the column
df['Engine'] = df['Engine'].apply(convert_engine_to_numeric)
df['Power'] = df['Power'].apply(convert_engine_to_numeric)
 
```

```python
df['Engine'].fillna(df['Engine'].mean(),inplace=True)
df['Power'].fillna(df['Power'].mean(),inplace=True)
df['Seats'].fillna(df['Seats'].mode()[0],inplace=True)
```
