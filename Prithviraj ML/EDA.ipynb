1. Chi-Square Test of Independence
Formula:

Assumptions:

    Independence: The observations should be independent. Each individual must belong to only one category.
    Sample Size: The sample size should be sufficiently large. Typically, each expected frequency should be at least 5.
    Categorical Data: The data should be in the form of frequencies or counts of categorical data.

2. Chi-Square Goodness of Fit Test
Formula:

The formula is similar:

χ2=∑(Oi−Ei)2Eiχ² = \sum \frac{(O_i - E_i)²}{E_i}χ2=∑Ei​(Oi​−Ei​)2​

Where:

    OiO_iOi​ = Observed frequency for category iii
    EiE_iEi​ = Expected frequency for category iii (based on the expected distribution)

Assumptions:

    Independence: Each observation is independent of the others.
    Sample Size: Each expected frequency should be at least 5.
    Data Type: Data must be categorical.

Example for Goodness of Fit:

Suppose you have a die, and you roll it 60 times and observe the following frequencies:
Face 	Observed Frequency
1 	10
2 	10
3 	15
4 	5
5 	10
6 	10

Expected Frequency for each face (assuming fair die) = 606=10\frac{60}{6} = 10660​=10

Using χ2χ²χ2 formula, compute and compare against a Chi-Square distribution to see if it fits a fair die.
