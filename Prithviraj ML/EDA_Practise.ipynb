```python
df1 = df.drop(columns=['embarked','class','deck','alive'],axis=1)
```

```python
df1['age'].fillna(df['age'].mean(),inplace=True)
df1['embark_town'].fillna(df1['embark_town'].mode()[0],inplace = True)
```

```python
df1.isnull().sum()
```

```python
df1.drop_duplicates(inplace=True)
```

```python
#new_feature
def cat(age):
    if age < 18:
        return "Young"
    elif age < 30:
        return "Adult"
    elif age < 55:
        return "Middle Age" 
    else:
        return "Senior"

df1["Age_category"]  =  df1['age'].apply(cat)
```

```python
#data visualization
numerical = df1.select_dtypes(include=['int','float']).columns.tolist()
category = df1.select_dtypes(exclude=['int','float']).columns.tolist()
numerical,category
```

```python
#Univariate
plt.figure(figsize=(12, 8))
df1.hist( figsize=(12, 8), bins=10, alpha=0.7, color='blue', edgecolor='black')
plt.title(f'Histograms')
plt.xlabel('column')
plt.ylabel('Frequency')

plt.tight_layout()  # Adjust spacing between plots
plt.show()
```
